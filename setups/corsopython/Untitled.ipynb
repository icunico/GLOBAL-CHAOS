```python
#!/usr/bin/env python3

class Book():
    def __init__(self, title, author="", read=False, prequel=None, sequel=None):
        self.title = title
        self.author = author
        self.read = read
        self.prequel = prequel
        self.sequel = sequel
    
    def mark_as_read(self):
        self.read = True
        
        # in case this is not the first book
        # set also the prequels as read
        if self.prequel != None:
            self.prequel.mark_as_read()
        
    def set_sequel(self,nextbook):
        self.sequel = nextbook
        nextbook.prequel = self
        
    def set_prequel(self,lastbook):
        self.prequel = lastbook
        lastbook.sequel = self
        
        
if __name__ == "__main__":
        book1 = Book("The Fellowship of the Ring", author="Tolkien")
        book2 = Book("The Two Towers", author="Tolkien")
        book3 = Book("The Return of the King", author="Tolkien")



        #Here you can see that book1 is not marked as read
        print(book1.read)

        book2.set_prequel(book1)
        book2.set_sequel(book3)

        # I set the third book as read
        book3.mark_as_read()

        # Here you can see that book2 is now marked as read
        # since it's the prequel of book3. The same holds for
        # book1 since it's marked as read in turn, when book2
        # is marked as read
        print(book2.read)
```

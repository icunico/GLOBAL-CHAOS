```python
class Book():
    def __init__(self, title, author, read=False, prequel=None, sequel=None):
        self.title=title
        self.author=author
        self.read=read
        self.prequel=prequel
        self.sequel=sequel

    def mark_as_read(self):
        self.read=True

        
        if self.prequel != None:
            self.prequel.mark_as_read()

book1=Book('Shantaram', 'Giovanni')
book2=Book('India', 'Paolo',True)
        
        
```

```python
print(book1)
```

```python
print(book1.read)
```

```python
book2.mark_as_read()
```

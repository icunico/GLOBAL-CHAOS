```python
class animals():
    prop='Mario'
    def __init__(self, name, age, color, owner=prop):
        self.name=name
        self.age=age
        self.color=color
        self.owner= owner

    def display(self):
        print(f"This {self.name} is {self.age} years old and {self.color} animal. Its owner is {self.owner}.") 

dog1=animals('dog',3,'black','Flavio')
```

```python
frog1=animals('frog', 10, 'green')
```

```python
dog1.display()
```
